```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Tratamento dos dados

## Analise numérica

```python
df = pd.read_csv('./melting-point/train.csv')
```

```python
desc = df.describe()
desc.T
```

## Remoção de colunas zeradas

```python
rm = []
for col in desc:
    if desc[col]['min'] == desc[col]['max']:
        rm.append(col)

print(rm)
df_without_zeros = df.drop(columns=rm, axis=1)
df_without_zeros
```

## Remoção SMILES e id

```python
df_o_wz_ws = df_without_zeros.drop(['id', 'SMILES'], axis=1)
df_o_wz_ws
```

## Remoção de duplicatas

```python
df_o_wz_ws = df_o_wz_ws.drop_duplicates()
df_o_wz_ws
```

```python
df_o_wz_ws.to_csv("./melting-point/train_o_wz_ws.csv", index=False)
```

## Escalonamento

```python
melting_point={}
melting_point["data"] = df_o_wz_ws.drop(['Tm'], axis=1).values
melting_point["target"] = df_o_wz_ws["Tm"].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

X = ss.fit_transform(melting_point['data'])
Y = melting_point['target']
```

```python
df_o_wz_ws_scaled = pd.DataFrame(data=X)
df_o_wz_ws_scaled = df_o_wz_ws_scaled.join(pd.DataFrame(data=Y, columns=['Tm']))
df_o_wz_ws_scaled.to_csv("./melting-point/train_o_scaled.csv", index=False)
```
